```python
import pandas as pd
import os
import numpy as np
```

```python
raw = pd.read_csv("aita_clean.csv")
# Create the vars to avoid key error
raw["text"] = ''*len(raw)
raw['age'] = np.zeros(len(raw))
raw['gender'] = np.zeros(len(raw))
```

```python
# Number of total examples
print(str(len(raw)))
```

```python
def concatenate(df):
    df["text"] = df["title"] + " " + df["body"]
```

```python
def extract_age(df):
    df["age"] = df["text"].str.extract(r'(?:(?:I|[Mm]y|[Mm]ine|[Mm]e|AITA|I\'m|I\'m\sa)[\s]*(?:\(|\[)(?:F|M|f|m)?(?:,|\s)*(\d+)(?:,|\s)*(?:F|M|f|m)?(?:\)|\]))|(?:(?:I\'m|I)[\sa-zA-Z]*(\d+))')
    
    
```

```python
def extract_gender(df):
    df["gender"] = df["text"].str.extract(r'(?:(?:I|[Mm]y|[Mm]ine|[Mm]e|AITA|I\'m|I\'m\sa)[\sa-zA-Z]*(?:\(|\[)\d+(?:,|\s)*(F|M|f|m)?(?:\)|\]))|(?:(?:I|[Mm]y|[Mm]ine|[Mm]e|AITA|I\'m|I\'m\sa)[\sa-zA-Z]*(?:\(|\[)(F|M|f|m)?(?:,|\s)*\d+(?:\)|\]))')

    
```

```python
concatenate(raw)
extract_age(raw)
extract_gender(raw)
```

```python
print(raw["age"].notnull().sum())
print(raw["gender"].notnull().sum())
```

```python
for index,row in raw.iterrows():
    if pd.notnull(raw.at[index,"age"]):
        if pd.isnull(raw.at[index,"gender"]) == True:
            print(index, row)
```

```python
raw.to_csv("aita_processed.csv")
```
