```python
import pandas as pd
import os
import numpy as np
```

```python
raw = pd.read_csv("aita_clean.csv")
```

```python
# Number of total examples
print(str(len(raw)))
```

```python
def concatenate(df):
    df["text"] = df["title"] + " " + df["body"]
```

```python
def extract_age(df):
    df["age"] = ""
    df["a1"] = ""
    df["a2"] = ""
    
    df["a1"] = df["text"].str.extract(r'(?:(?:I|[Mm]y|[Mm]ine|[Mm]e|AITA|I\'m|I\'m\sa)[\s]*(?:\(|\[)(?:F|M|f|m)?(?:,|\s)*(\d+)(?:,|\s)*(?:F|M|f|m)?(?:\)|\]))')
    df["a2"] = df["text"].str.extract(r'(?:I\'m|I\'m\sa|I\sam|I)[\s]*(\d+)')
    #(?:(?:I|[Mm]y|[Mm]ine|[Mm]e|AITA|I\'m|I\'m\sa)[\s]*(?:\(|\[)(?:F|M|f|m)?(?:,|\s)*(\d+)(?:,|\s)*(?:F|M|f|m)?(?:\)|\]))|(?:(?:I\'m|I)[\sa-zA-Z]*(\d+))
    for index,row in df.iterrows():
        if pd.notnull(df.at[index,"a1"]):
            df.at[index,"age"] = df.at[index,"a1"]
        elif pd.notnull(df.at[index,"a2"]):
            df.at[index,"age"] = df.at[index,"a2"]
            
    df["age"] = df["age"].replace({'': np.nan})
    del df["a1"]
    del df["a2"]
            
```

```python
def extract_gender(df):
    df["gender"] = ""
    df["g1"] = ""
    df["g2"] = ""
    df["g3"] = ""
    df["g4"] = ""
    
    df["g1"] = df["text"].str.extract(r'(?:I|AITA|[Mm]y|[Mm]ine|[Mm]e)[\s]*(?:\(|\[)\d+(?:,|\s)*(F|M|f|m)(?:\)|\])')
    df["g2"] = df["text"].str.extract(r'(?:I|AITA|[Mm]y|[Mm]ine|[Mm]e)[\s]*(?:\(|\[)(?:(F|M|f|m)(?:,|\s)*\d+)(?:\)|\])')
    df["g3"] = df["text"].str.extract(r'(?:I\'m|I\'m\sa)[\sa-zA-Z]*(?:\(|\[)?\d+(?:,|\s)*(F|M|f|m)(?:\)|\])?')
    df["g4"] = df["text"].str.extract(r'(?:I\'m|I\'m\sa)[\sa-zA-Z]*(?:\(|\[)?(F|M|f|m)(?:,|\s)*\d+(?:\)|\])?')
    #(?:I|AITA|[Mm]y|[Mm]ine|[Mm]e)[\s]*(?:\(|\[)(?:\d+(?:,|\s)*(F|M|f|m)|(F|M|f|m)(?:,|\s)*\d+)(?:\)|\])|(?:I\'m|I\'m\sa)[\sa-zA-Z]*(?:\(|\[)?(?:\d+(?:,|\s)*(F|M|f|m)|(F|M|f|m)(?:,|\s)*\d+)(?:\)|\])?
    for index,row in df.iterrows():
        if pd.notnull(df.at[index,"g1"]):
            df.at[index,"gender"] = df.at[index,"g1"]
        elif pd.notnull(df.at[index,"g2"]):
            df.at[index,"gender"] = df.at[index,"g2"]
        elif pd.notnull(df.at[index,"g3"]):
            df.at[index,"gender"] = df.at[index,"g3"]
        elif pd.notnull(df.at[index,"g4"]):
            df.at[index,"gender"] = df.at[index,"g4"]
    
    df["gender"] = df["gender"].replace({'': np.nan})
    del df["g1"]
    del df["g2"]
    del df["g3"]
    del df["g4"]
            
```

```python
def change_labels(df):
    for index,row in df.iterrows():
        if df.at[index,"verdict"] == "everyone sucks":
            df.at[index,"verdict"] = "asshole"
        if df.at[index,"verdict"] == "no assholes here":
            df.at[index,"verdict"] = "not the asshole"
```

```python
concatenate(raw)
extract_age(raw)
extract_gender(raw)
change_labels(raw)
print(raw)
```
